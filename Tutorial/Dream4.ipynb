```python
import pandas as pd 
from sklearn.preprocessing import RobustScaler
import os
import torch
import LTGN

Robust_scaler = RobustScaler()
```

```python
dream4path1 = r'F:\GitDemo\LTGN\data\dream4\insilico_size10_1_timeseries.tsv'
p1 = r'F:\GitDemo\LTGN\data\dream4\insilico_size10_1_goldstandard.tsv'
```

```python
df = pd.read_csv(dream4path1, sep='\t', skip_blank_lines=True, header=None)
Gene_label = df.iloc[:1]
Gene_label = Gene_label.iloc[:, 1:]
Gene_label = Gene_label.values.tolist()
Gene_label = Gene_label[0]
df = df.iloc[1:]
df = df.iloc[:, 1:]
df = df.astype(float)
df_array = df.to_numpy()
Robust_scaler.fit(df_array)
df_array = Robust_scaler.transform(df_array)

dream4_gene = Gene_label
gene_to_id = {name: idx for idx, name in enumerate(dream4_gene)}

df = pd.read_csv(p1, sep='\t', skip_blank_lines=True, header=None)
df.columns = ['source','target','label']
df = df[df['label'] == 1]
df['source_id'] = df['source'].map(gene_to_id)
df['target_id'] = df['target'].map(gene_to_id)
edge_list = df[['source_id', 'target_id']].values.tolist()

dream4_data = [df_array]
dream4_gs = [edge_list]
```

```python
# parameter and data
threshold, seed = (0.25, 373)
args = (0.25, 373)
data_dream_0 = df_array
gs = edge_list

filename = 'dream4' + str(0) + 'RB' + 'threshold' + str (threshold) + 'seed '+ str(seed) + '.txt'
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
```

```python
orr,tpr,fpr,pres,ac,gsl,edgee = LTGN.onlyData(
                                         data = data_dream_0, 
                                         gs = gs, 
                                         threshold = threshold, 
                                         lr = 1e-3, 
                                         epoch = 1, 
                                         # Please modify the training epoch.
                                         datasetname = 'dream4'+'rb'+'th'+str(threshold)+str(0)+'seed'+str(seed),
                                         ln = 30,
                                         seed = seed,
                                         genename = dream4_gene,
                                         relationship = 'c',
                                         ref = 'pc',
                                         device = device)
```
